```python
import pandas as pd
from openpyxl import load_workbook
```

```python
df = pd.read_excel('ReporteGeneralSpring2027.xls', skiprows=6)
print(f'Número de registros {len(df)}')
```

```python
columnas_proceso = ['CI', 'Celular','Nombre','Sponsor','Empleador', 'Posicion Laboral', 'Correo', 'Fecha Inscripcion', 'Precio']
```

```python
df['Retirado motivo'].unique()
```

```python
df['Nombre'] = df["Nombre"].str.strip().str.lower()
```

# Pacticipantes **Reales**  y **Activos**

```python
mask_estudiantes_activos = df['Estatus Participante'] == "Activo"
mask_no_prueba = ~(df['Nombre'].str.contains('prueba'))
perfiles_fake = ["veronica alejandra ollage velastegí", "camila prueba", "emily prueba prueba prueba", "ashly prueba"]
mask_perfiles_reales = ~(df['Nombre'].isin(perfiles_fake))
```

```python
df_activos  = df.loc[mask_estudiantes_activos & mask_perfiles_reales & mask_no_prueba]
print(f"Participantes activos => {len(df_activos)}")
```

# PRUEBA INGLES ESCRITA

```python
sin_prueba_escrita = df_activos['Calificacion Escrita'] == 'Sin Calificacion'
con_prueba_escrita = df_activos['Calificacion Escrita'] != 'Sin Calificacion'
print(f"Activos => {len(df_activos)}")
print(f'Participantes SIN prueba de inglés escrita => {sin_prueba_escrita.sum()}')
print(f'Participantes CON prueba de inglés escrita => {con_prueba_escrita.sum()}')
print(f'Cuadre pruebas => {sin_prueba_escrita.sum() + con_prueba_escrita.sum()}')
```

# PRUEBA DE INGLES ORAL

```python
sin_prueba_oral = df_activos['Cualitativa'] == "Sin Nota"
con_prueba_oral = df_activos['Cualitativa'] != "Sin Nota"
print(f"Activos => {len(df_activos)}")
print(f'Participantes SIN prueba de inglés oral => {sin_prueba_oral.sum()}')
print(f'Participantes CON prueba de inglés oral => {con_prueba_oral.sum()}')
print(f'Cuadre pruebas => {sin_prueba_oral.sum() + con_prueba_oral.sum()}')
```

# Sin capturar precio

```python
mask_precio_no_capturado = df_activos['Precio'].str.split(' ').str[-1].astype('float') == 0
df_precios_no_capturados = df_activos.loc[mask_estudiantes_activos & mask_precio_no_capturado, columnas_proceso].sort_values(by='Fecha Inscripcion')
print(f"Participantes sin precio capturado => {len(df_precios_no_capturados)}")
```

# Parcticipantes inscritos el día de ayer

```python
yesterday = (pd.Timestamp.now() - pd.to_timedelta('1 day'))
yesterday.date()
```

```python
yesterday = pd.Timestamp(yesterday).normalize()
yesterday
```

```python
mask_just_yesterday = df['Fecha Inscripcion'] == yesterday
```

```python
print(f"Estudiantes inscritos el día de ayer {len(df.loc[mask_estudiantes_activos & mask_just_yesterday, columnas_proceso])}")
```

```python
participantes_inscritos_ayer = df_activos.loc[mask_estudiantes_activos & mask_perfiles_reales & mask_just_yesterday, columnas_proceso]
participantes_inscritos_ayer
```

```python
str_yesterday = yesterday.strftime('%d-%m-%Y')
str_yesterday
```

```python
nombre_archivo = "Inscriptos_Spring2027_" + str_yesterday + ".xlsx"
with pd.ExcelWriter(nombre_archivo) as writer:
    participantes_inscritos_ayer.to_excel(writer)
print("Excel guardado")
```

```python
wb = load_workbook(nombre_archivo)
ws = wb.active


for col in ws.columns:
   max_lenght = max(len(str(cell.value)) for cell in col if cell.value)
   ws.column_dimensions[col[0].column_letter].width = max_lenght + 2

wb.save(nombre_archivo)
```

# Prueba agendadas
