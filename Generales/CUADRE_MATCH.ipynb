```python
import pandas as pd
from openpyxl import load_workbook
```

```python
columnas_para_colocar_en_base_match = ['Sponsor', 'Estatus Participante','Universidad', 'Carrera', 'Nombre', 'CI', 'Celular', 'Correo', 'Empleador', 'Posicion Laboral', 'Fecha Inscripcion']
df = pd.read_excel(io="ReporteGeneralSpring2027.xls", usecols=columnas_para_colocar_en_base_match, skiprows=6)
df = df.sort_values(by='Sponsor')
print(f'Número de registros {len(df)}')
```

```python
perfiles_fake = ["EMILY PRUEBA PRUEBA PRUEBA", "Ashly Prueba"]
mask_activos = df['Estatus Participante'] == 'Activo'
mask_perfiles_reales = ~(df['Nombre'].isin(perfiles_fake))
```

```python
df_match = df.loc[mask_activos & mask_perfiles_reales, columnas_para_colocar_en_base_match]
```

```python
print(f"Participantes activos => {len(df_match)}")
```

```python
fecha_corte = pd.Timestamp.now().normalize().strftime('%d-%m-%Y')
nombre_archivo = 'Participantes_MATCH' + fecha_corte + '.xlsx'
```

```python
with pd.ExcelWriter(nombre_archivo) as writer:
    df_match.to_excel(writer)
    print("Excel guardado")

wb = load_workbook(nombre_archivo)
ws = wb.active


for col in ws.columns:
   max_lenght = max(len(str(cell.value)) for cell in col if cell.value)
   ws.column_dimensions[col[0].column_letter].width = max_lenght + 2

wb.save(nombre_archivo)
```
